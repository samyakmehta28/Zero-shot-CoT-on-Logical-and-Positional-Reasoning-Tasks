```python
import json

# Open the file and read the contents
with open('my_log/family_relationship_puzzle.json', 'r') as file:
    data = json.load(file)

# Print the contents of the file
count=0
for item in data:
    if item['prediction'] != item['Ground_Truth']:
        count=count+1
        print("-------------------------------------------------------------------")
        print(f"Prompt: {item['prompt']}")
        print(f"Response: {item['response']}")
        print(f"Prediction: {item['prediction']}")
        print(f"Ground Truth: {item['Ground_Truth']}\n")
print(count)
```

```python
import json

# Open the file and read the contents
with open('my_log/family_relationship_puzzle.json', 'r') as file:
    data = json.load(file)
total_count=0
correct_count=0
for item in data:
    if "row" in item['prompt'] or "line" in item['prompt']:
        #print(item['prompt'])
        if item['prediction'] != item['Ground_Truth'] and "two rows" in item['prompt']:
            print("-------------------------------------------------------------------")
            print(f"Prompt: {item['prompt']}")
            print(f"Response: {item['response']}")
            print(f"Prediction: {item['prediction']}")
            print(f"Ground Truth: {item['Ground_Truth']}\n")
            correct_count=correct_count+1
            break
        total_count=total_count+1
print(correct_count, total_count)
print((correct_count/total_count)*100)
```
